### Importing Libraries

```python
import yfinance as yf
import pandas as pd
import numpy as np
import os
```

### Data Fetching

```python
# 1. Tickers (Companys name)
tickers = ['RELIANCE.NS', 'TCS.NS', 'INFY.NS']

print("Fetching Data from yahoo finance, please wait... This may take a few seconds.")

# 2. Downloading data (Past 1 year)
data = yf.download(tickers, period="1y")

# 3. Data check
if data.empty:
    print("Error: Data nahi mila! Apna internet connection check karein.")
else:
    print("Success! Data fetched.")
    display(data.head())
```

```python
folder_path = "../Dataset"
os.makedirs(folder_path, exist_ok=True)

# File saving path
file_path = f"{folder_path}/stock_data.csv"

# Saving data to CSV
data.to_csv(file_path)

print(f"Data Saved Successfully: {file_path}")
```
